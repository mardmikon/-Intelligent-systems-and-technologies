```python
import numpy as np
import pandas as pd
import time
import math
import matplotlib.pyplot as plt
from geneticalgorithm import geneticalgorithm
from scipy.optimize import differential_evolution
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# **B. Решение задачи оптимизации многоэкстремальной функции, используя классический алгоритм оптимизации и эволюционный алгоритм.**


**Оптимизируемая функция Бута**

> Формула: f(x, y) = ((x + 2 * y - 7) ^ 2 + (2 * x + y - 5) ^ 2)

> Глобальный минимум: f(1,3) = 0

> Метод поиска: -10 <= x, y <= 10




```python
def Boot(var):
    return ((var[0] + 2 * var[1] - 7) ** 2 + (2 * var[0] + var[1] - 5) ** 2)
```

**100 прогонов для Генетического алгоритма на функции Бута**

```python
bestFunctionBoot = []
print("Старт оптимизации.")
for step in range(100):
  varbound = np.array([[-10,10]]*2)
  model = geneticalgorithm(function=Boot, 
                          dimension=2, 
                          variable_type='real', 
                          variable_boundaries=varbound)

  print(f"Прогон - {step}.")
  timeModelStart = time.time()
  model.run()
  bestFunctionBoot.append(model.best_function)
  timeModelStop = time.time()
  print(f"Время прогона: {timeModelStop-timeModelStart}")
print("Завершение оптимизации.")
```

Вычисление математическое ожидания и дисперсии для финального
значения функции

```python
print(f"expectation = {np.mean(bestFunctionBoot)}, dispersion =  {np.var(bestFunctionBoot)}")
```

Так как значение дисперсии является крайне малым, то алгоритм будет сходится

**100 прогонов для алгоритма дифференциальной эволюции на функции Бута**

```python
def cb(x, convergence):
    progress.append(x)
    progress_val.append(Boot(x))

bestFunctionDE = []
print("Старт оптимизации.")
for step in range(100):
  progress = []
  progress_val = []

  print(f"Прогон - {step}.")
  timeModelStart = time.time()
  bounds = np.array([[-10,10]]*2)
  result = differential_evolution(Boot, bounds, callback=cb)

  progress = np.array(progress)
  progress_val = np.array(progress_val)

  fig = plt.figure()
  ax = fig.add_subplot(2,1,2)
  ax.plot(progress_val)
  plt.show()
  bestFunctionDE.append(result.fun)
  timeModelStop = time.time()
  print(f"Время прогона: {timeModelStop-timeModelStart}")
print("Завершение оптимизации.")
```

Вычисление математическое ожидания и дисперсии для финального значения функции

```python
print(f"expectation = {np.mean(bestFunctionDE)}, dispersion =  {np.var(bestFunctionDE)}")
```

Так как значение дисперсии является крайне малым, то алгоритм будет сходится

**Вывод:** алгоритм дифференциальной эволюции продемострировал лучшие показатели матожидания и дисперсии по сравнению с генетическим алгоритмом

Так как алгоритм дифференциальной эволюции показал лучший резлуьтат, чем генетический алгоритм, то он и будет применен при поиске оптимальных значений.

```python
headers = ["word_freq_make", "word_freq_address", "word_freq_all", "word_freq_3d", "word_freq_our", "word_freq_over", "word_freq_remove", "word_freq_internet", "word_freq_order", "word_freq_mail", "word_freq_receive",
                "word_freq_will", "word_freq_people", "word_freq_report", "word_freq_addresses", "word_freq_free", "word_freq_business", "word_freq_email", "word_freq_you", "word_freq_credit", "word_freq_your", "word_freq_font",
                "word_freq_000", "word_freq_money", "word_freq_hp", "word_freq_hpl", "word_freq_george", "word_freq_650", "word_freq_lab", "word_freq_labs", "word_freq_telnet", "word_freq_857", "word_freq_data",
                "word_freq_415", "word_freq_85", "word_freq_technology", "word_freq_1999", "word_freq_parts", "word_freq_pm", "word_freq_direct", "word_freq_cs", "word_freq_meeting", "word_freq_original", "word_freq_project",
                "word_freq_re", "word_freq_edu", "word_freq_table", "word_freq_conference", "char_freq_;", "char_freq_(", "char_freq_[", "char_freq_!", "char_freq_$", "char_freq_#", "capital_run_length_average", "capital_run_length_longest","capital_run_length_total"]

dataset = pd.read_csv('spambase.data', names=headers)
```

```python
_standartScaler = StandardScaler()
x_train, x_test, y_train, y_test = train_test_split(_standartScaler.fit_transform(dataset.iloc[:,1:-1]),
                                                    dataset.iloc[:,-1],
                                                    test_size=0.5,
                                                    shuffle=True,
                                                    stratify=dataset.iloc[:,-1],
                                                    random_state=100)
_standartScaler = StandardScaler()
```

**SVM-классификатор**


> В работе 5-6 для моего набора данных SVM классификатор получил следующие лучшие параметры:


d48a9797-b897-470a-9bd7-2c8c5de214f1.png

```python
def cb(x, convergence):
    progress.append(x)
    progress_val.append(Boot(x))

def SVMOptimizer(var):
    svm = SVC(kernel='rbf', C=var[0], gamma=var[1])
    svm.fit(x_train, y_train)
    x_test_pred = svm.predict(x_test)
    accuracy = accuracy_score(y_test, x_test_pred)
    return -accuracy

progress = []
progress_val = []
bounds = np.array([[1,100], [0.001, 1.0]])
result = differential_evolution(SVMOptimizer, bounds, callback=cb)

progress = np.array(progress)
progress_val = np.array(progress_val)

fig = plt.figure()
ax = fig.add_subplot(2,1,2)
ax.plot(progress_val)
plt.show()
```

```python
svmOrigin = SVC(kernel='rbf')
svmOrigin.fit(x_train, y_train)
predOrigin = svmOrigin.predict(x_test)
accuracyOrigin = accuracy_score(y_test, predOrigin)
accuracyData = [accuracyOrigin, -result.fun]
axis = ['Исход', 'Дифференциальная эволюция']
plt.bar(axis, accuracyData, color=['black', 'black'])
print(f"Точность:")
print(f"Исходные тестовые данные - {accuracyOrigin}")
print(f"Данные после алгоритма дифференциальной эволюции - {-result.fun}")
```

**KNN-классификатор**


> В работе 5-6 для моего набора данных KNN классификатор получил следующие лучшие параметры:

b680ce2a-72cc-4396-8743-49b61a3fca76.png

```python
def cb(x, convergence):
    progress.append(x)
    progress_val.append(Boot(x))

def KNNOptimizer(var):
    knn = KNeighborsClassifier(metric='euclidean', 
              weights='distance', 
              n_neighbors=int(var[0]),
              leaf_size=int(var[1]))
    
    knn.fit(x_train, y_train)
    x_test_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test, x_test_pred)
    return -accuracy

progress = []
progress_val = []
bounds = np.array([[1,500], [1,500], [1, 20]])
result = differential_evolution(KNNOptimizer, bounds, callback=cb)

progress = np.array(progress)
progress_val = np.array(progress_val)

fig = plt.figure()
ax = fig.add_subplot(2,1,2)
ax.plot(progress_val)
plt.show()
```

```python
knnOrigin = KNeighborsClassifier(metric='euclidean', weights='distance')
knnOrigin.fit(x_train, y_train)
predOrigin = knnOrigin.predict(x_test)
accuracyOrigin = accuracy_score(y_test, predOrigin)
accuracyData = [accuracyOrigin, -result.fun]
axis = ['Исход', 'Дифференциальная эволюция']
plt.bar(axis, accuracyData, color=['black', 'black'])
print(f"Точность:")
print(f"Исходные тестовые данные - {accuracyOrigin}")
print(f"Данные после алгоритма дифференциальной эволюции - {-result.fun}")
```

**RF-классификатор**


> В работе 5-6 для моего набора данных RF классификатор получил следующие лучшие параметры:

fb4d8dfc-db41-4089-b0c3-20ccdc867937.png

```python
def cb(x, convergence):
    progress.append(x)
    progress_val.append(Boot(x))

def RFOptimizer(var):
    rf = RandomForestClassifier(criterion='entropy',
                                bootstrap=False,
                                max_features='sqrt',
                                n_estimators=int(var[0]),
                                max_depth = int(var[1]),
                                min_samples_split = int(var[2])/100.0,
                                min_samples_leaf = int(var[3]))
    
    rf.fit(x_train, y_train)
    x_test_pred = rf.predict(x_test)
    accuracy = accuracy_score(y_test, x_test_pred)
    return -accuracy

progress = []
progress_val = []
bounds = np.array([[1,80], [1,80], [1,100],[1,80]])
result = differential_evolution(RFOptimizer, bounds, callback=cb)

progress = np.array(progress)
progress_val = np.array(progress_val)

fig = plt.figure()
ax = fig.add_subplot(2,1,2)
ax.plot(progress_val)
plt.show()
```

```python
rfOrigin = RandomForestClassifier(criterion='entropy', bootstrap=False, max_features='sqrt')
rfOrigin.fit(x_train, y_train)
predOrigin = rfOrigin.predict(x_test)
accuracyOrigin = accuracy_score(y_test, predOrigin)
accuracyOrigin = accuracy_score(y_test, predOrigin)
accuracyData = [accuracyOrigin, -result.fun]
axis = ['Исход', 'Дифференциальная эволюция']
plt.bar(axis, accuracyData, color=['black', 'black'])
print(f"Точность:")
print(f"Исходные тестовые данные - {accuracyOrigin}")
print(f"Данные после алгоритма дифференциальной эволюции - {-result.fun}")
```

**Вывод:** Исходные значения были успешно оптимизированы благодаря алгоритму дифференциальной эволюции, что привело к увеличению значения точноти на тестовых данных. Однако стоит отметить: в случае RF-классификатора наблюдается небольшое ухудшение показателя точности
